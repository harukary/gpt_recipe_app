```python
from bs4 import BeautifulSoup
import requests,re

# HTMLページを取得する
url = "https://www.lettuceclub.net/recipe/dish/22978/"
# url = "https://park.ajinomoto.co.jp/recipe/card/709401/"
response = requests.get(url)

# BeautifulSoupを使用してテキストを抽出する
soup = BeautifulSoup(response.content, "html.parser")
text = soup.get_text().replace('\n\n','')

# テキストを出力する
print(text)
```

```python
end = '他のレシピをもっと見る'

# 文字列stringから、substring以降を削除する
new_text = text[:text.find(end)]

# 結果を表示する
print(new_text) # "ab"
print(len(new_text))
```
